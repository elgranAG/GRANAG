```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('astronauta.jpg')
b, g, r = cv2.split(imagen)

# Calcular el histograma original
hist_original = cv2.calcHist([b], [0], None, [256], [0, 256])

# Desplazar el histograma
val_des = 50
img_des = cv2.add(b, val_des)

# Fijar el rango de valores entre 0 y 256
img_des = np.clip(img_des, 0, 255).astype(np.uint8)

# Calcular el histograma desplazado
hist_des = cv2.calcHist([img_des], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title('Histograma Original')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(hist_original)

plt.subplot(1, 2, 2)
plt.title('Shifted Histogram')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(hist_des)

plt.tight_layout()
plt.show()

# Display the original and shifted images (optional)
cv2.imshow('Original Imagen', imagen)
cv2.imshow('Shifted Imagen', img_des)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

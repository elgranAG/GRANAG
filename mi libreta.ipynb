## CÃ“DIGO DE DESENFOQUE


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('galaxia.jpg', cv2.IMREAD_COLOR)
imagenbw = cv2.imread('galaxia.jpg', cv2.IMREAD_GRAYSCALE)
imagenRGB = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
cv2.imwrite('galaxia.jpg', imagenRGB)

edge_kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
imagen1 = cv2.filter2D(imagenbw, -1, edge_kernel)

gaussiam_kernel2 = cv2.getGaussianKernel(500, 0)
imagen2 = cv2.filter2D(imagen, -1, gaussiam_kernel2)
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.title('imagen original')
plt.imshow(imagen)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('deteccion de bordes')
plt.imshow(imagen1, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('desenfoque')
plt.imshow(imagen2)
plt.axis('off')

plt.tight_layout()
plt.show()
```
